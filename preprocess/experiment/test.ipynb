```python
import matplotlib.pyplot as plt

# 设置列的数量，根据您提供的图片，这里我们假设有6列
columns = 6

# 初始化图表的 x 轴坐标，这里我们使用从 0 到 5 的整数
x = range(columns)

# 初始化图表的 y 轴坐标，这里我们使用从 100 开始，每列增加 0.5°
y = [100 + i * 0.5 for i in x]

# 创建图表
plt.figure(figsize=(10, 5))  # 设置图表的大小

# 绘制每一列的线
for i in range(columns):
    plt.plot(x, [y[i]] * columns, marker='o')  # 使用圆点标记每个数据点

# 设置图表标题和坐标轴标签
plt.title('Sample Chart with Incrementing Degrees')
plt.xlabel('Columns')
plt.ylabel('Values (°)')

# 显示网格
plt.grid(True)

# 显示图表
plt.show()
```

```python
import matplotlib.pyplot as plt

# 设置列的数量，这里我们使用6列作为示例
columns = 6

# 初始化图表的 x 轴坐标
x = range(columns)

# 初始化图表的 y 轴坐标，从100开始，每列增加0.5°
y = [100 + i * 0.5 for i in range(columns)]

# 创建图表
plt.figure(figsize=(8, 1))  # 设置图表的大小

# 绘制每一列的线，这里我们使用一个非常小的点来表示
for i in range(columns):
    plt.plot(x, [y[i]] * len(x), marker='o', markersize=1, color='black', linestyle='-')

# 隐藏坐标轴
plt.axis('off')

# 显示图表
plt.show()
```

```python
import matplotlib.pyplot as plt

# 设置列的数量，这里我们有7列，从100到160.0，每列增加0.5
columns = 120
max_degree = 160.0
start_degree = 100

# 初始化图表的 x 轴坐标
x = range(columns)

# 初始化图表的 y 轴坐标，每列数值相同，从100开始，每列增加0.5
y = [start_degree + i * 0.5 for i in range(columns)]

# 创建图表
plt.figure(figsize=(10, 2), facecolor='white')  # 设置图表的大小和底色为白色

# 绘制每一列的垂直线
for i in range(columns):
    plt.plot(x, [y[i]] * columns, color='black', linewidth=1.0)

# 添加数值标签
for i in range(columns):
    plt.text(i, y[i], f'{y[i]:.1f}', ha='center', va='bottom')

# 隐藏坐标轴
plt.axis('off')

# 显示图表
plt.show()
```
